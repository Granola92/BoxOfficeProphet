```python
#Initiate currency converter function and read dataframe
c = CurrencyConverter()
df = pd.read_csv('df_movie.csv', index_col = ['Name'])


# Function to convert currency symbol to 3-letter code
def convert_currency_symbol_to_code(currency_symbol):
    symbol_to_code = {
        '$': 'USD',
        '¬£': 'GBP',
        '‚Ç¨': 'EUR',
        '‚Çπ': 'INR',
        '‚Ç©': 'KRW',
        '¬•': 'JPY'
        # Add more currency symbols and their corresponding codes as needed
    }
    if isinstance(currency_symbol, pd.Series):
        return currency_symbol.map(symbol_to_code).fillna('Unknown')
    return symbol_to_code.get(currency_symbol, 'Unknown')

# Create 'Budget Currency Code' column with 3-letter codes
df['Budget Currency Code'] = convert_currency_symbol_to_code(df['Budget Currency'])

# Function to convert to USD
# Modify the conversion function to handle missing rates
def convert_to_usd(row):
    if row['Budget Currency'] == '$':
        return row['Estimated Budget']
    elif row['Budget Currency'] == '¬£':
        return c.convert(row['Estimated Budget'], 'GBP', 'USD', date=date(2013, 3, 21))
    elif row['Budget Currency'] == '‚Ç¨':
        return c.convert(row['Estimated Budget'], 'EUR', 'USD')
    elif row['Budget Currency'] == '‚Çπ':
        return c.convert(row['Estimated Budget'], 'INR', 'USD')
    elif row['Budget Currency'] == '‚Ç©':
        return c.convert(row['Estimated Budget'], 'KRW', 'USD')
    elif row['Budget Currency'] == '¬•':
        return c.convert(row['Estimated Budget'], 'JPY', 'USD')

# Apply the conversion function to create the 'Budget Converted' column
df['Budget Converted'] = df.apply(convert_to_usd, axis=1) 

# Display the updated DataFrame
print(df)
```
